```python
data=[1,2,3,4,5,6,7,8,8,7,7,6,6]
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='r',alpha=.8)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
